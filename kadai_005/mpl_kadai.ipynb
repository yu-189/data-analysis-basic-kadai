```python
import matplotlib.pyplot as plt
import numpy as np

# 複数個のグラフを表示する領域を作成します。Figure(描画領域全体)とAxes(ひとつひとつのプロットを描く領域)を作成するコードを記述する。
# 描画領域全体にサブプロットを一つ生成するコードです
fig, ax = plt.subplots()
```

```python
fig, axes = plt.subplots(2, 3)

y0 = [1, 2, -5, 2]
axes[0,0].plot(y0)

x1 = np.linspace(0, 10, 100)
y1 = 2 + 2 * np.sin(2 * x1)
axes[0,1].plot(x1,y1)

x2 = np.random.normal(15,5,2000)
#ヒストグラムをプロット
axes[1,0].hist(x2,bins=30)

x3 = 4 + np.random.normal(0,2,60)
y3 = 4 + np.random.normal(0,2,60)
axes[1,1].scatter(x3,y3)

plt.show()
```
